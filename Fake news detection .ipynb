```python
import numpy as np
import matplotlib.pyplot as plt

# Algorithms and their metrics for each class
classifiers = ['Logistic Regression', 'Decision Tree Classifier', 'Gradient Boosting Classifier', 'Random Forest Classifier']
precision_fake = [0.99, 1.00, 1.00, 0.99]
recall_fake = [0.98, 1.00, 0.99, 0.99]
f1_fake = [0.99, 1.00, 1.00, 0.99]

precision_true = [0.98, 1.00, 0.99, 0.99]
recall_true = [0.99, 1.00, 1.00, 0.99]
f1_true = [0.99, 1.00, 1.00, 0.99]

# Setting the width of bars
bar_width = 0.2
index = np.arange(len(classifiers))

# Plotting bars for Fake class
plt.bar(index - bar_width, precision_fake, bar_width, label='Precision (Fake)', color='blue')
plt.bar(index, recall_fake, bar_width, label='Recall (Fake)', color='lightblue')
plt.bar(index + bar_width, f1_fake, bar_width, label='F1-score (Fake)', color='skyblue')

# Plotting bars for True class
plt.bar(index - bar_width, precision_true, bar_width, label='Precision (True)', color='green', hatch='/')
plt.bar(index, recall_true, bar_width, label='Recall (True)', color='lightgreen', hatch='/')
plt.bar(index + bar_width, f1_true, bar_width, label='F1-score (True)', color='lime', hatch='/')

# Adding titles and labels
plt.xlabel('Classifiers')
plt.ylabel('Score')
plt.title('Precision, Recall, and F1-Score for Each Classifier')
plt.xticks(index, classifiers, rotation=45)
plt.ylim(0.95, 1.01)  # Limiting y-axis for better visualization
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Classifier names
classifiers = ['LR', 'DTC', 'GBC', 'RFC']

# Metrics for each classifier and class (Fake and True)
precision_fake = [0.99, 1.00, 1.00, 0.99]
recall_fake = [0.98, 1.00, 0.99, 0.99]
f1_fake = [0.99, 1.00, 1.00, 0.99]

precision_true = [0.98, 1.00, 0.99, 0.99]
recall_true = [0.99, 1.00, 1.00, 0.99]
f1_true = [0.99, 1.00, 1.00, 0.99]

# Plotting the line plots
plt.figure(figsize=(10, 6))

# Fake class plots
plt.plot(classifiers, precision_fake, marker='o', linestyle='-', label='Precision (Fake)', color='blue')
plt.plot(classifiers, recall_fake, marker='o', linestyle='-', label='Recall (Fake)', color='cyan')
plt.plot(classifiers, f1_fake, marker='o', linestyle='-', label='F1-score (Fake)', color='skyblue')

# True class plots
plt.plot(classifiers, precision_true, marker='s', linestyle='--', label='Precision (True)', color='green')
plt.plot(classifiers, recall_true, marker='s', linestyle='--', label='Recall (True)', color='lightgreen')
plt.plot(classifiers, f1_true, marker='s', linestyle='--', label='F1-score (True)', color='lime')

# Adding titles and labels
plt.xlabel('Classifiers')
plt.ylabel('Score')
plt.title('Classification Report Metrics for Each Classifier')
plt.ylim(0.95, 1.01)  # Set y-axis limits for better comparison
plt.legend(loc='lower right')
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```
